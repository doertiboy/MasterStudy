```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%store -r acc_dict
%store -r jerk_values
%store -r features_dict
person_names = ["p1", "p2", "p3", "p4", "p5", "p6", "p7", "p8", "p9", "p10", "p11", "p12", "p13", "p14", "p15", "p16"]
unique_colors = sns.color_palette('tab20')
```

# Scatter Plots

```python
def plotFeatures(feature_x, feature_y):
    for i, person_name in enumerate(person_names):
        features = features_dict[person_name]
        x = features[feature_x]
        y = features[feature_y]
        gender = features['gender']

        color = 'red' if gender == 'female' else 'blue'
        plt.scatter(x, y, color=color)
```

```python
import matplotlib.pyplot as plt

def plotFeatures(subplot_list):
    num_subplots = len(subplot_list)
    fig, axs = plt.subplots(nrows=1, ncols=num_subplots, figsize=(5 * num_subplots, 5))

    for i, (feature_x, feature_y) in enumerate(subplot_list):
        c = 0
        for person_name in person_names:
            features = features_dict[person_name]
            x = features[feature_x]
            y = features[feature_y]
            gender = features['gender']
            marker = 'x' if gender == 'male' else 'o'
            axs[i].scatter(x, y, marker=marker, label=person_name, color=unique_colors[c])
            c+=1
        axs[i].set_xlabel(feature_x)
        axs[i].set_ylabel(feature_y)

    plt.tight_layout()
    plt.show()
```

## Jerk Vector / Time

```python
subplot_list = [('mean_jerk_vector', 'time'), ('mean_jerk_1_vector', 'time_1'), ('mean_jerk_2_vector', 'time_2'), ('mean_jerk_3_vector', 'time_3')]
plotFeatures(subplot_list)
```

```python
subplot_list = [('mean_jerk_y', 'time'), ('mean_jerk_1_y', 'time_1'), ('mean_jerk_2_y', 'time_2'), ('mean_jerk_3_y', 'time_3')]
plotFeatures(subplot_list)
```

## Jerk / Peaks

```python
subplot_list = [('mean_jerk_vector', 'pos_peaks'), ('mean_jerk_1_vector', 'pos_peaks_1'), ('mean_jerk_2_vector', 'pos_peaks_2'), ('mean_jerk_3_vector', 'pos_peaks_3')]
plotFeatures(subplot_list)
```

## Time / Peaks

```python
subplot_list = [('time', 'pos_peaks'), ('time_1', 'pos_peaks_1'), ('time_2', 'pos_peaks_2'), ('time_3', 'pos_peaks_3')]
plotFeatures(subplot_list)
```
